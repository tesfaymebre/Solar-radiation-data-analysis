```python
# check for negative values in the dataset

numeric_cols = benin_df.select_dtypes(exclude=['object']).columns

benin_df[numeric_cols].lt(0).sum()
```

```python
def detect_outliers_iqr(df, columns):
    outlier_summary = {}
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        outlier_summary[col] = len(outliers)
    return outlier_summary

outlier_counts = detect_outliers_iqr(benin_df, numeric_cols)
print(outlier_counts)
```

```python
# Convert the negative values in ['GHI', 'DNI', 'DHI'] to absolute values
benin_df[['GHI', 'DNI', 'DHI']] = benin_df[['GHI', 'DNI', 'DHI']].abs()

# detect outliers again
outlier_counts = detect_outliers_iqr(benin_df, numeric_cols)
print(outlier_counts)
```

```python
# outliers specific to sensor columns
sensor_cols = ['ModA', 'ModB', 'WS', 'WSgust']
sensor_outliers = detect_outliers_iqr(benin_df, sensor_cols)
print(sensor_outliers)
```

```python
# Use box plots or histograms to visualize the distribution and identify anomalies.
for col in sensor_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=benin_df[col])
    plt.title(f'Boxplot for {col}')
    plt.show()
```
